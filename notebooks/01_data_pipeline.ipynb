```python
import pandas as pd
import numpy as np
import yfinance as yf

print("pandas:", pd.__version__)
print("numpy:", np.__version__)
print("yfinance:", yf.__version__)
print("All libraries loaded successfully.")
```

```python
ticker = yf.Ticker("AAPL")

info = ticker.info

print("Company:", info.get("longName"))
print("Sector:", info.get("sector"))
print("Industry:", info.get("industry"))
print("Current Price:", info.get("currentPrice"))
print("Market Cap:", info.get("marketCap"))
print("P/E Ratio:", info.get("trailingPE"))
print("EV/EBITDA:", info.get("enterpriseToEbitda"))
print("Revenue:", info.get("totalRevenue"))
```

```python
history = ticker.history(period="5y")

print(history.head())
print("\nShape:", history.shape)
print("\nDate range:", history.index[0], "to", history.index[-1])
```

```python
income_stmt = ticker.financials
balance_sheet = ticker.balance_sheet
cash_flow = ticker.cashflow

print("Income Statement columns:", income_stmt.shape)
print("Balance Sheet columns:", balance_sheet.shape)
print("Cash Flow columns:", cash_flow.shape)

print("\nIncome Statement rows:")
print(income_stmt.index.tolist())
```

```python
rows_we_need = [
    'Total Revenue',
    'Gross Profit',
    'EBITDA',
    'Operating Income',
    'Net Income',
    'Basic EPS'
]

snapshot = income_stmt.loc[rows_we_need]

snapshot.columns = [str(col)[:10] for col in snapshot.columns]

print(snapshot.to_string())
```

```python
import os

os.makedirs("../data/raw/AAPL", exist_ok=True)

history.to_csv("../data/raw/AAPL/price_history.csv")
income_stmt.to_csv("../data/raw/AAPL/income_statement.csv")
balance_sheet.to_csv("../data/raw/AAPL/balance_sheet.csv")
cash_flow.to_csv("../data/raw/AAPL/cash_flow.csv")

print("Saved successfully.")
```

```python
import pandas as pd

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"}

sp500_table = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies", storage_options={"User-Agent": headers["User-Agent"]})
sp500_df = sp500_table[0]

print(sp500_df.shape)
print(sp500_df[['Symbol', 'Security', 'GICS Sector']].head(10))
```

```python
tickers = sp500_df['Symbol'].tolist()

print("Total tickers:", len(tickers))
print("\nFirst 10:", tickers[:10])
print("Last 10:", tickers[-10:])

sp500_df.to_csv("../data/raw/sp500_tickers.csv", index=False)
print("\nSaved ticker list.")

print(sp500_df.columns.tolist())
```

```python
import time
import os

failed_tickers = []

for i, ticker_symbol in enumerate(tickers):
    try:
        ticker_obj = yf.Ticker(ticker_symbol)
        
        os.makedirs(f"../data/raw/{ticker_symbol}", exist_ok=True)
        
        ticker_obj.history(period="5y").to_csv(f"../data/raw/{ticker_symbol}/price_history.csv")
        ticker_obj.financials.to_csv(f"../data/raw/{ticker_symbol}/income_statement.csv")
        ticker_obj.balance_sheet.to_csv(f"../data/raw/{ticker_symbol}/balance_sheet.csv")
        ticker_obj.cashflow.to_csv(f"../data/raw/{ticker_symbol}/cash_flow.csv")
        
        print(f"[{i+1}/503] {ticker_symbol} done")
        time.sleep(1.5)
        
    except Exception as e:
        print(f"[{i+1}/503] {ticker_symbol} FAILED — {e}")
        failed_tickers.append(ticker_symbol)
        continue

print("\nFinished.")
print("Failed tickers:", failed_tickers)
```

```python
import os

folders = [f for f in os.listdir("../data/raw") if os.path.isdir(f"../data/raw/{f}")]

print(f"Total company folders: {len(folders)}")
print(f"Sample folders: {folders[:5]}")

with open("../data/raw/pull_log.txt", "w") as f:
    f.write(f"S&P 500 data pulled on: {pd.Timestamp.now()}\n")
    f.write(f"Total companies: {len(folders)}\n")
    f.write(f"Failed tickers: {failed_tickers}\n")

print("Log saved.")
```
